# **Bivariate Analysis**

```python
# Visualization 1: swarm plot of trends in murder and attempt to murder over years
plt.figure(figsize=(12, 6))
sns.swarmplot(x='year', y='murder', label='Murder', data=df)
sns.swarmplot(x='year', y='attempt_to_murder', label='Attempt to Murder', data=df)
plt.title('Trends in Murder and Attempt to Murder Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()

# Visualization 2: Violin plot of different types of crimes
crime_columns = ['dacoity', 'riots', 'burglary', 'cheating', 'assult_on_women']
plt.figure(figsize=(14, 8))
sns.violinplot(data=df[crime_columns], inner='quartile')
plt.title('Distribution of Different Types of Crimes')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.show()

# Visualization 3: Heatmap of correlation between crime types
crime_correlation = df[crime_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(crime_correlation, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Between Different Types of Crimes')
plt.show()
#Bar plot
plt.figure(figsize=(20, 6))
sns.barplot(x='state_ut', y='total_ipc_crimes', data=df.sort_values(by='total_ipc_crimes', ascending=False))
plt.title('Total IPC Crimes by State/Union Territory')
plt.xlabel('State/Union Territory')
plt.ylabel('Total IPC Crimes')
plt.show()
```

Visualization 1 - Swarm Plot of Murder and Attempt to Murder Trends Over Years:

The swarm plot depicts the distribution of murder and attempt to murder counts
across different years.
It reveals any patterns, outliers, or variations in these crime types over the specified time period.
Observations can be made regarding the frequency and trends in murder and attempted murder incidents over the years.

Visualization 2 - Violin Plot of Different Types of Crimes:

The violin plot illustrates the distribution of selected crime types (dacoity, riots, burglary, cheating, assault on women).
It provides a comprehensive view of the spread and concentration of each crime category, aiding in the identification of potential outliers and variations.
Comparisons between crime types become visually apparent through the quartile representation.

Visualization 3 - Heatmap of Correlation Between Crime Types:

The heatmap showcases the correlation matrix between different types of crimes.
It visually identifies relationships and dependencies between crime categories, highlighting potential associations.
Positive or negative correlations between crime types are easily discernible, providing valuable insights into the dataset's internal dynamics.
Bar Plot - Total IPC Crimes by State/Union Territory:

The bar plot presents a clear comparison of total IPC (Indian Penal Code) crimes across different State/Union Territories.
It allows for the identification of regions with higher or lower total crime rates, providing a snapshot of the overall crime scenario.
Variations in crime rates among states become apparent through the visual representation.

# Multivariate Analysis

```python
# Create a new DataFrame with selected columns for the catplot
catplot_data = df[['dacoity', 'riots', 'burglary', 'cheating', 'assult_on_women', 'state_ut']]

# Melt the DataFrame to long format
catplot_data_long = catplot_data.melt(id_vars='state_ut', var_name='Crime Type', value_name='Count')

# Create a catplot
plt.figure(figsize=(14, 8))
sns.catplot(x='Crime Type', y='Count', hue='state_ut', data=catplot_data_long, kind='bar', height=6, aspect=2)
plt.title('Catplot of Different Types of Crimes Across States/UTs')
plt.xlabel('Crime Type')
plt.ylabel('Count')
plt.show()
```

The catplot, visualized using Seaborn, effectively presents a comparative analysis of different types of crimes across various State/Union Territories. The transformation of the dataset into long format and the use of categorical plotting provide valuable insights into crime distribution and variations.

Key Observations:

Crime Type Comparison:

The catplot compares the counts of different crime types (dacoity, riots, burglary, cheating, assault on women) across State/Union Territories.
Vertical bars represent the counts for each crime type, allowing for a visual assessment of prevalence.

State/UT Comparison:

The hue parameter distinguishes bars by different colors, representing individual State/Union Territories.
This enables a direct comparison of crime distribution within each region.

Categorical Representation:

The categorical plot format is conducive to understanding both the distribution of each crime type and the variations among different regions.

```python
x_variable = 'total_ipc_crimes'
y_variable = 'rape'

# Create a hexbin plot
plt.figure(figsize=(10, 8))
sns.jointplot(x=x_variable, y=y_variable, data=df, kind='hex')
plt.suptitle(f'Hexbin Plot of {x_variable} vs. {y_variable}', y=1.02)
plt.show()
```

The hexbin plot, visualized using Seaborn's jointplot function, provides a bivariate representation of the relationship between two variables: total_ipc_crimes (X-variable) and rape (Y-variable). The hexbin plot is a useful tool for visualizing the density of points in the scatter plot, particularly when dealing with a large number of data points.

Key Observations:

Density Representation:

The hexbin plot effectively represents the density of points in the scatter plot by aggregating them into hexagonal bins.
The intensity of color within each hexagon indicates the concentration of data points.

Correlation Patterns:

By observing the hexbin plot, insights into the correlation patterns between total_ipc_crimes and rape can be derived.
The plot allows for the identification of areas with higher or lower density, suggesting potential correlation strength.

Distribution of Data:

The hexbin plot provides a visual summary of the distribution of data points along the X and Y axes, offering a quick overview of their joint distribution.


```python
multivariate_columns = ['preparation_and_assembly_for_dacoity', 'robbery', 'burglary', 'theft','state_ut', 'other_theft', 'riots', 'criminal_breach_of_trust']

# Create a parallel coordinates plot
plt.figure(figsize=(14, 8))
parallel_coordinates(df[multivariate_columns], class_column='state_ut', colormap='rainbow')
plt.title('Parallel Coordinates Plot of Selected Crimes Across States/UTs')
plt.show()
```

The parallel coordinates plot, generated for selected crime variables across different State/Union Territories, offers a comprehensive visual representation of multivariate data. This type of plot is particularly effective for comparing and understanding the patterns of multiple variables simultaneously.

Key Observations:

Multivariate Comparison:

The parallel coordinates plot allows for the simultaneous comparison of multiple crime variables, including preparation_and_assembly_for_dacoity, robbery, burglary, theft, other_theft, riots, and criminal_breach_of_trust.

State/UT Distinctions:

Each line in the plot represents a State/Union Territory, and the connections between points on different axes illustrate the values of each crime variable for a specific region.
Distinctive patterns and variations in crime profiles among different regions become visually apparent.

Relative Magnitudes:

The plot provides insights into the relative magnitudes of different crime variables within each State/UT.
Patterns of higher or lower values for specific crimes in certain regions are observable.
