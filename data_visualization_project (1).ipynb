```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
from google.colab import autoviz

```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.info()      #details of file
```

```python
#drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()   #for checking null values
```

```python
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
#change datatypes
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
# I have change the column name.
df.rename(columns= {'Marital_Status':'Shaadis'})
```

```python
df.describe(include=['object', 'int','float'])
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender',hue='Gender',data=df)
plt.title('Distribution of customers by marital status')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.violinplot(df, x='State', y='Age')
plt.title("Violinplot")
plt.show()
```

```python
plt.figure(figsize=(20, 5))
sns.violinplot(df, x='Age Group', y='Orders',)
plt.title("Violinplot")
plt.show()
```

```python
df['Amount'].plot(kind="hist")
```

```python
# Question 1: Overall trend in Diwali sales over the age groups
plt.figure(figsize=(10, 5))
sns.lineplot(x='Age Group', y='Zone', data=df)
plt.title('Diwali Sales Trend Over the Years')
plt.xlabel('Age group')
plt.ylabel('State')
plt.show()

```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Age Group', data=df)
plt.title('Distribution of Customers by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='State', data=df)
plt.title('Geographic Distribution of Customers by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(18, 8))
sns.boxplot(x='Age', y='Amount', data=df)
plt.title('Correlation Between Age Groups and Amount')
plt.xlabel('Age')
plt.ylabel('Amount (in $)')
plt.xticks(rotation=45)
plt.show()
```

```python
marital_status_counts = df['Age Group'].value_counts()
plt.figure(figsize=(8, 8))

plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Age Groups')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

```python
heatmap_data = df.pivot_table(index='State', columns=['Marital_Status', 'Age Group'], values='User_ID', aggfunc='count')
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Distribution of Customers by State, Marital Status, and Age Group')
plt.xlabel('Marital Status & Age Group')
plt.ylabel('State')
plt.show()
```

```python

# Create the scatter plot
plt.figure(figsize=(19, 6))
sns.scatterplot(x='Occupation', y='Age', data=df)
plt.title('Scatter Plot of Occupation vs. Age')
plt.xlabel('Occupation')
plt.ylabel('Age')
plt.show()
```

```python

# Create the hexbin plot
plt.figure(figsize=(10, 6))
sns.jointplot(x='Orders', y='Amount', data=df, kind='hex', cmap='YlGnBu')
plt.title('Hexbin Plot of Orders vs. Amount')
plt.xlabel('Orders')
plt.ylabel('Amount (in $)')

# Show the hexbin plot
plt.show()
```
